## 6. Displaying only months 

```python
df['published_month']=df['published_at'].str[5:7]
df[['published_month']]
```

## 7. Replacing the numbers in the column published_month

```python
months={'01':'Jan','02':'Feb','03':'Mar','04':'Apr','05':'May','06':'Jun','07':'Jul','08':'Aug','09':'Sep','10':'Oct','11':'Nov','12':'Dec'}
months
```

```python
df['published_month']=df['published_month'].map(months)
df['published_month']
```
